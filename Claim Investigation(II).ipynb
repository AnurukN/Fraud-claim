```python
for col in id_col_name:
    all_claim_data_step_2[col] = all_claim_data_step_2[col].astype('str')

for col in date_col_name:
    all_claim_data_step_2[col] = pd.to_datetime(all_claim_data_step_2[col], format="%Y%m%d", errors='coerce')
    
for col in category_col_name:
     all_claim_data_step_2[col] = all_claim_data_step_2[col].astype('category')
```
